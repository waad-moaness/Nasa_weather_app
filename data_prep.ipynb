```python
import pandas as pd
from functools import reduce

ALLSKY_SFC_SW_DW = "nasa_data/ALLSKY_SFC_SW_DWN.csv"
max_temp = "nasa_data/max_temp.csv"
max_WS2M = "nasa_data/max_WS2M.csv"
min_temp = "nasa_data/min_temp.csv"
min_WS2M = "nasa_data/min_WS2M.csv"
prectotcorr = "nasa_data/prectotcorr.csv"
RH2M = "nasa_data/RH2M.csv"
temp = "nasa_data/temp.csv"
WS2M = "nasa_data/WS2M.csv"


df1 = pd.read_csv(ALLSKY_SFC_SW_DW, skiprows=9)
df2 = pd.read_csv(temp, skiprows=9)
df3 = pd.read_csv(max_temp, skiprows=9)
df4 = pd.read_csv(min_temp, skiprows=9)
df5 = pd.read_csv(prectotcorr, skiprows=9)
df6 = pd.read_csv(RH2M, skiprows=9)
df7 = pd.read_csv(WS2M, skiprows=9)
df8 = pd.read_csv(min_WS2M, skiprows=9)
df9 = pd.read_csv(max_WS2M, skiprows=9)

dfs = [df1, df2, df3, df4 , df5, df6 ,df7,df8 , df9]

nasa_df = reduce(lambda left, right: pd.merge(left, right, on=["LAT","LON","YEAR","DOY"]), dfs)

nasa_df.to_csv("merged24.csv", index=False)

nasa_df
```
